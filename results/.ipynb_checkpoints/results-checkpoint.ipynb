## Results

### Loading necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

### Loading the dataset

```python
df = pd.read_csv("training_step_data.csv")
```

### Normalize column names

```python
df.columns = [c.strip().lower().replace(" ", "_") for c in df.columns]
print("Detected columns:", df.columns.tolist())

def find_col(keywords):
    for kw in keywords:
        for col in df.columns:
            if kw in col:
                return col
    return None

episode_col = find_col(["episode", "ep", "step"])
reward_col = find_col(["reward", "return", "score"])
energy_col = find_col(["energy", "power", "load", "demand", "consumption"])
loss_col = find_col(["loss", "error", "cost"])
```

### Configure plotting style

```python
sns.set_theme(style="whitegrid")
plt.rcParams.update({
    "font.family": "serif",
    "font.serif": ["Times New Roman"],
    "font.size": 9,
    "axes.titlesize": 10,
    "axes.labelsize": 9,
    "legend.fontsize": 8,
    "figure.dpi": 300,
    "savefig.dpi": 600
})
```

### Reward Convergence Plot

```python
plt.figure(figsize=(4.2, 2.8))
sns.lineplot(data=df, x=episode_col, y=reward_col, color="#1f77b4", linewidth=1.3, label="Total Reward")
plt.title("Reward Convergence Over Episodes")
plt.xlabel("Episode")
plt.ylabel("Total Reward")
plt.legend(loc="upper right", frameon=False)
plt.tight_layout()
plt.savefig("reward_convergence_compact.png", bbox_inches="tight")
plt.show()
print("Explanation: Reward increases steadily, confirming learning stability and effective coordination among agents.\n")
```

### Energy Consumption Trend

```python
plt.figure(figsize=(7, 4))
sns.lineplot(data=df, x=episode_col, y=energy_col, color="#2ca02c", linewidth=1.3, label="Energy Consumption")
plt.title("Energy Consumption Over Episodes")
plt.xlabel("Episode")
plt.ylabel("Energy (kWh)")
plt.legend(frameon=False)
plt.tight_layout()
plt.savefig("energy_trend_compact.png", bbox_inches="tight")
plt.show()
print("Explanation: Energy consumption stabilizes, indicating improved grid balance and adaptive demand response.\n")
```
